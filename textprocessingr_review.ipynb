```python
import bs4
import urllib.request as url
```

```python
data={'review':[]}

for i in range(1,500):
    response=url.urlopen('https://www.flipkart.com/apple-iphone-13-starlight-128-gb/product-reviews/itmc9604f122ae7f?pid=MOBG6VF5ADKHKXFX&lid=LSTMOBG6VF5ADKHKXFXQGX7PK&aid=overall&certifiedBuyer=false&sortOrder=MOST_HELPFUL&page={}'.format(i))
    page=bs4.BeautifulSoup(response)
    review=page.find_all('div',class_='t-ZTKy')
    
    for j in range(len(review)):
        data['review'].append(review[j].text)
    
```

```python
import pandas as pd
```

```python
df=pd.DataFrame(data)
```

```python
import nltk
nltk.download('punkt')
nltk.download('stopwords')
nltk.download('wordnet')
nltk.download('omw-1.4')
```

```python
import string
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
df.review[0].split()
```

```python
table = str.maketrans('','',string.punctuation)
```

```python
df.review[1].translate(table)
```

```python
df.review[0].translate(table)
```

```python
df.review[2].translate(table)
```

```python
# Remove punctuations
for i in range(len(df['review'])):
    df.review[i] = df.review[i].lower().translate(table)
```

```python
print(df.review)
```

```python
print(df['review'])
```

```python
# word tokenization
for i in range(len(df['review'])):
    df.review[i] = word_tokenize(df.review[i])
```

```python
print(df['review'])
```

```python
englishStopwords = stopwords.words("english")
```

```python
words = []
for tokens in df['review']:
    word = []
    for i in range(len(tokens)):
        if tokens[i] not in englishStopwords:
            word.append(tokens[i])
    words.append(word)
```

```python
print(words)
```

```python
wnet = WordNetLemmatizer()
```

```python
wnet.lemmatize('watching')
```

```python
wnet.lemmatize('bought','v')
```

```python
for i in range(len(words)):
    for j in range(len(words[i])):
        words[i][j] = wnet.lemmatize(words[i][j], 'v')
```

```python
print(words)
```

```python
# convert in sentense
for i in range(len(words)):
    words[i] = " ".join(words[i])
```

```python
print(words)
```

```python
words
```

```python
tfidf = TfidfVectorizer()
```

```python
tfidf.fit(words)
```

```python
print(tfidf.vocabulary_)
```
